# Matrices and loops

```python
size(600, 600)
color_mode(HSB, 360, 100, 100)
background(0, 0, 100)

squares_per_line = 5
gap = 20
margin = width * 0.15
empty_space = (margin * 2) + (gap * (squares_per_line - 1))
side = (width - empty_space) / squares_per_line

stroke(0, 0, 100)

for i in range(0, squares_per_line):
    for j in range(0, squares_per_line):
        x = margin + (i * (gap + side))
        y = margin + (j * (gap + side))

        color_value = (360 / (squares_per_line ** 2)) * (i + j) * random(0, 3)
        
        no_fill()
        stroke_weight(3)
        fill(color_value, 70, 90)
        rect(x, y, side, side)

        inner_square_gap = 15
        if random(5) > 2.5:
            stroke_weight(random_int(2, 15))
            rect(x + inner_square_gap, y + inner_square_gap, side - (inner_square_gap * 2), side - (inner_square_gap * 2))
```
