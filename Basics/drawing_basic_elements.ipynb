# Drawing basic elements

Exploring color and primitives

Ref: [py5coding.org](https://py5coding.org/tutorials/intro_to_py5_and_python_02_drawing_2d_primitives.html).

```python
size(500, 500)
background("#ffffff")

canvas_size = 500
square_side = canvas_size / 2
x = y = (canvas_size - square_side) / 2

stroke_join(ROUND)

no_stroke()
fill("#FF0000")
rect(x, y, square_side, square_side)

stroke("#ffffff")
stroke_weight(15)
fill("#FF9900")
rect(x / 2, y / 2, square_side, square_side)

stroke("#000000")
no_fill()
stroke_weight(15)
rect(x * 1.5, y * 1.5, square_side, square_side)
```

```python
size(500, 500)
background("#113388")

stroke("#ffffff")
no_fill()

stroke_weight(2)
point(250, 10)
stroke_weight(5)
point(13, 65)
stroke_weight(3)
point(405, 90)
stroke_weight(2)
point(100, 33)

no_stroke()
fill("#ffffff")
ellipse(250, 150, 150, 150)
```

```python
size(500, 500)
background("#444444")

fill("#f3f3f3")
no_stroke()

# triangle points
point_1 = 100, 300
point_2 = 400, 300
point_3 = 150, 100

# quad points
point_4 = 150, 300
point_5 = 380, 100
point_6 = 245, 100
point_7 = 300, 300

triangle(*point_1, *point_2, *point_3)

fill("#000000")
quad(*point_4, *point_5, *point_6, *point_7)
```
