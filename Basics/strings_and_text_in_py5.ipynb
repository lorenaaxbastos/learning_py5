```python
size(500, 500)
background("#004477")
fill("#fff")
stroke("#0099ff")
stroke_weight(3)
# Py5Font.list() to get a list of fonts installed

razor = "Never attribute to malice that which is adequately explained by stupidity."
elephant = create_font("Elephant", 20)

text_size(20)
text_leading(20)
text(razor, 40, 100, 200, 200)

text_font(elephant)
text(razor, 250, 100, 220, 200)

pristina = create_font("Pristina", 30)
text_align(CENTER)
text_font(pristina)
text_leading(35)
text(razor, 100, 250, 300, 500)

hanlons = "- Hanlon's razor"
text_align(LEFT)
book_antiqua = create_font("Book Antiqua", 15)
text_font(book_antiqua)
text(hanlons, 350, 450)

# A line on either side of our text!
line(340, 425, 340, 460)
line(340 + text_width(hanlons) + 20, 425, 340 + text_width(hanlons) + 20, 460)
```
