```python
size(600, 600)
background("#004477")

x = width / 2
y = height / 2
rect_w = 0.75 * width
rect_h = 0.75 * height

no_stroke()
fill("#ffffff")
rect_mode(CENTER)
rect(x, y, rect_w, rect_h)
```

```python
size(600, 740)
background("#DAA520")
no_fill()
stroke("#ffffff")
stroke_weight(3)

xco = 400
yco = 440
center_x = width / 2
center_y = height / 2


line(center_x, height / 3, xco, yco)
ellipse(center_x, center_y, width / 11, height / 14)
ellipse(center_x, center_y, width / 19, height / 22)
line(xco, yco, width - xco, yco)
line(width - xco, yco, center_x, height / 3)
ellipse(center_x, center_y, width / 5, height / 12)
```

```python
size(600, 700)
background("#FDBCB4")
stroke("#660000")
no_fill()

def degree_to_rad(degree):
    return degree * PI / 180

radius_inner = 300
radius_outer = 400
radius_most_inner = 200
radius_most_outer = 500
radius_pie = 200
center_x = width / 2
center_y = height / 2

stroke_weight(3)
arc(
    center_x, center_y,
    radius_most_inner, radius_most_inner,
    degree_to_rad(-180), degree_to_rad(0)
)

stroke_weight(9)
arc(
    center_x, center_y,
    radius_inner, radius_inner,
    degree_to_rad(-90), degree_to_rad(90)
)

stroke_weight(6)
arc(
    center_x, center_y,
    radius_outer, radius_outer,
    degree_to_rad(0), degree_to_rad(180)
)

stroke_weight(12)
arc(
    center_x, center_y,
    radius_most_outer, radius_most_outer,
    degree_to_rad(0), TAU
)

no_stroke()
fill("#660000")
arc(
    center_x, center_y,
    radius_pie, radius_pie,
    PI / 4, PI * 3 / 4,
    PIE
)
```

```python
size(600, 600)

# Color pallete
dark_blue = "#112438"
dark_turquoise = "#005F73"
light_yellow = "#E9D8A6"
yellow = "#EE9B00"
dark_yellow = "#CA6702"
orange = "#BB3E03"
dark_red = "#9B2226"

background(dark_blue)
stroke("#112438")
stroke_weight(4)

# Arcs
center_x = width / 2
center_y = height * 2.5 / 4

radius = 550
fill(light_yellow)
arc(
    center_x, center_y,
    radius, radius,
    (TAU * 3 / 4) + HALF_PI / 8, (TAU * 3 / 4) + HALF_PI / 2.5,
    PIE
)

radius = 400
fill(orange)
arc(
    center_x, center_y,
    radius, radius,
    PI, TAU,
    PIE
)

fill(orange)
arc(
    center_x, center_y,
    radius, radius,
    PI + HALF_PI / 2, TAU,
    PIE
)

fill(yellow)
arc(
    center_x, center_y,
    radius, radius,
    (TAU * 3 / 4) + HALF_PI / 8, TAU,
    PIE
)

fill(yellow)
arc(
    center_x, center_y,
    radius, radius,
    (TAU * 3 / 4) + (HALF_PI * 2 / 3), TAU,
    PIE
)

radius = 250
fill(dark_red)
arc(
    center_x, center_y,
    radius, radius,
    PI, (TAU * 3 / 4) + HALF_PI / 8,
    PIE
)

fill(dark_yellow)
arc(
    center_x, center_y,
    radius, radius,
    (TAU * 3 / 4) + HALF_PI / 8, TAU,
    PIE
)

fill(dark_turquoise)
arc(
    center_x, center_y,
    radius, radius,
    0, PI,
    PIE
)

radius = 100
fill(dark_blue)
arc(
    center_x, center_y,
    radius, radius,
    0, TAU,
)
```
